```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

df = pd.read_csv('epa-sea-level.csv')

# Create scatter plot
x = df['Year']
y = df['CSIRO Adjusted Sea Level']

fig, ax = plt.subplots()
plt.scatter(x, y, marker='.')

# Create first line of best fit
xy = np.arange(1880, 2051, 1)
res1 = linregress(x, y)
line1 = [res1.slope*xi + res1.intercept for xi in xy]
plt.plot(xy, line1, color = 'orange', label="Fitting Line", linewidth=1)

# Create second line of best fit
df_y2k = df.loc[(df['Year'] >= 2000)]
x2 = df_y2k['Year']
y2 = df_y2k['CSIRO Adjusted Sea Level']

xy2 = np.arange(2000, 2051, 1)
res2 = linregress(x2, y2)
line2 = [res2.slope*xi + res2.intercept for xi in xy2]
plt.plot(xy2, line2, color = 'red', label="Fitting Line", linewidth=1)

# Add labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
```
